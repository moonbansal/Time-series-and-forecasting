```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from scipy.interpolate import LSQUnivariateSpline

boston = pd.read_csv('/Users/moonbansal/Downloads/4-1/TIME SERIES/boston-housing.csv')

boston.head()
```

```python
knots = np.percentile(boston['RM'], [25, 50, 75])

spline_custom_knots = LSQUnivariateSpline(boston['RM'], boston['MEDV'], t=knots, k=3)

t_fine = np.linspace(0, 100, 300)
#mean_spline_custom = spline_custom_knots(t_fine)

plt.figure(figsize=(8, 6))
plt.scatter(boston['RM'], boston['MEDV'], color='orange')
plt.plot(t_fine, spline_custom_knots, color='blue', linewidth=2)
#plt.scatter(knots)
plt.show()
```
